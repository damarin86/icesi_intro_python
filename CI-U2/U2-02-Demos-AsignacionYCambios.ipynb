### 2. Cambio aritmético

En ocasiones necesitamos realizar cambios sobre los valores numéricos, poara esto Python y otros lenguajes de programación cuentan con algunas opciones:

```python
# Incremento manual
valor = 0
print(f'Inicial: {valor}')
valor = valor + 1
print("Final: " + str(valor))
```

```python
# Incremento resumido
valor = 0
print("Inicial: " + str(valor))
valor += 1
print("Final: " + str(valor))
```

```python
# Decremento manual
valor = 1
print("Inicial: " + str(valor))
valor = valor - 1
print("Final: " + str(valor))
```

```python
# Decremento resumido
valor = 1
print("Inicial: " + str(valor))
valor -= 1
print("Final: " + str(valor))
```

```python
# Multiplicación manual y resumida
valor = 2
print("Inicial: " + str(valor))
valor = valor * 2
print("Final: " + str(valor))

valor = 2
print("Inicial: " + str(valor))
valor *= 2
print("Final: " + str(valor))
```

```python
# División manual y resumida
valor = 8
print("Inicial: " + str(valor))
valor = valor / 2
print("Final: " + str(valor))

valor = 8
print("Inicial: " + str(valor))
valor /= 2
print("Final: " + str(valor))
```

### 3. Otros

#### Concatenación

En el manejo de cadenas podemos realizar concatenación

```python
# Sólo tipo cadena str
cadenaA = "una frase"
cadenaB = "otra frase"

# Se queda todo pegado
cadenaConcatenada = cadenaA + cadenaB
print(cadenaConcatenada)

# Arreglamos el formato
cadenaConcatenada = cadenaA +" "+ cadenaB
print(cadenaConcatenada)

# Forma especial con "la cadena f"
cadenaConcatenada = f'{cadenaA} {cadenaB}'
print(cadenaConcatenada)
```

```python
# Múltiples tipos de dato
variableA = "una frase"
variableB = 10
variableC = 1.75
variableD = ["A","B","C"]
 
## ---> cadenaConcatenada = variableA + variableB + variableC + variableD # Error
cadenaConcatenada = str(variableA) + str(variableB) + str(variableC) + str(variableD)
print(cadenaConcatenada)

# Arreglamos el formato
cadenaConcatenada = str(variableA) +" "+ str(variableB) +" "+ str(variableC) +" "+ str(variableD)
print(cadenaConcatenada)

# Forma especial con "la cadena f"
cadenaConcatenada = f'{variableA} {variableB} {variableC} {variableD}'
print(cadenaConcatenada)
```
