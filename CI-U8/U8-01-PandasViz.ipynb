### 2. Tipos de gráficos
[[enlace]](#A00)

#### Gráfico histograma

```python
# Crear un DataFrame de ejemplo
data = {'Puntuaciones': [85, 90, 70, 65, 80, 95, 75, 60, 85, 90, 70, 65]}
df = pd.DataFrame(data)

# Crear el gráfico de histograma
df.plot.hist()
```

#### Gráfico de puntos

```python
# Crear un DataFrame de ejemplo
data = {'w': [10, 15, 12, 8, 20, 14, 16],
        'h': [5, 8, 6, 4, 10, 7, 9]}
df = pd.DataFrame(data)

# Crear el gráfico de dispersión
df.plot.scatter(x='w', y='h')
```

#### Gráfico de línea

```python
data = {'x': [1, 2, 3, 4, 5], 'y': [2, 4, 6, 8, 10], 'z': [2, 7, 6, 12, 10] }
df = pd.DataFrame(data)
df.plot.line()
```

#### Gráfico barras

```python
data = {'x': ['A', 'B', 'C', 'D', 'E'], 'y': [3, 6, 2, 7, 4]}
df = pd.DataFrame(data)
df.plot.bar()
```

#### Gráfico barras apiladas

```python
data = {'x': ['A', 'B', 'C'], 'y1': [3, 6, 2], 'y2': [4, 2, 7]}
df = pd.DataFrame(data)
df.plot.bar(stacked=True)
```

#### Gráfico barras horizontales

```python
data = {'x': ['A', 'B', 'C', 'D'], 'y': [5, 3, 2, 6]}
df = pd.DataFrame(data)
df.plot.barh()
```

#### Gráfico áreas

```python
data = {'x': [1, 2, 7, 4, 5], 'y': [2, 4, 3, 8, 10]}
df = pd.DataFrame(data)
df.plot.area()
```

#### Gráfico de cajas y bigotes

```python
data = {'x': ['A', 'A', 'B', 'B', 'C'], 'y': [3, 4, 6, 8, 5]}
df = pd.DataFrame(data)
df.plot.box()
```

#### Gráfico torta

```python
data = {'x': ['A', 'B', 'C'], 'y': [40, 25, 35]}
df = pd.DataFrame(data)
df.plot.pie(y='y')
```

#### Combinación con otras librerías

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)
z = np.random.randn(100)

data = {'x': x, 'y': y, 'z': z}
df = pd.DataFrame(data)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['x'], df['y'], df['z'])
plt.show()
```

### 3. Cambio de color y estilos
[[enlace]](#A00)

```python
data = {'x': [1, 2, 3, 4, 5], 'y': [2, 4, 6, 8, 10], 'z': [1, 3, 6, 12, 18]}
df = pd.DataFrame(data)

df.plot(colormap='viridis')
# más colores en https://matplotlib.org/stable/tutorials/colors/colormaps.html
```

```python
data = {'x': [1, 2, 3, 4, 5], 'y1': [2, 4, 6, 8, 10], 'y2': [1, 3, 5, 7, 9]}
df = pd.DataFrame(data)
df.plot(color=['red', 'blue', 'green'])
```

```python
data = {'x': [1, 2, 3, 4, 5], 'y': [2, 4, 6, 8, 10]}
df = pd.DataFrame(data)

df.plot(style=['r--', 'bo'])
```

#### Opciones para el estilado

| Estilo | Descripción         |
|--------|---------------------|
| `'-'`  | Línea sólida        |
| `'--'` | Línea punteada      |
| `'-.'` | Línea guion-punto   |
| `':'`  | Línea de puntos     |
| `'.'`  | Puntos              |
| `','`  | Puntos pequeños     |
| `'o'`  | Círculos            |
| `'v'`  | Triángulos hacia abajo |
| `'^'`  | Triángulos hacia arriba |
| `'<'`  | Triángulos hacia la izquierda |
| `'>'`  | Triángulos hacia la derecha |
| `'s'`  | Cuadrados           |
| `'p'`  | Pentágonos          |
| `'*'`  | Estrellas           |
| `'+'`  | Cruces              |
| `'x'`  | Aspas               |
| `'D'`  | Diamantes           |
| `'d'`  | Diamantes pequeños  |
| `'h'`  | Hexágonos           |
| `'H'`  | Hexágonos grandes   |


```python
df.plot(style=['r-', 'bo', 'g+'])
```

```python
df.plot(style=['g--', 'rs'])
```

```python
df.plot(style=['b-.', 'y*'])
```

```python
df.plot(style=['k:', 'gv'])
```
