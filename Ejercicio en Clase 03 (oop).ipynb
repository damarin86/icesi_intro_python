```python
# Añade métodos llamados depositar y retirar a la clase CuentaBancaria que modifiquen el saldo de la cuenta.

class CuentaBancaria:
    def __init__(self, saldo):
        self.__saldo = saldo

    def obtener_saldo(self):
        return self.__saldo

    def depositar(self, cantidad):
        self.__saldo += cantidad

    def retirar(self, cantidad):
        self.__saldo -= cantidad

cuenta = CuentaBancaria(1000)
print(cuenta.obtener_saldo())
cuenta.depositar(500)
print(cuenta.obtener_saldo())
cuenta.retirar(200)
print(cuenta.obtener_saldo())
```

```python
# Añade un método llamado transferir a la clase CuentaBancaria que transfiera fondos de una cuenta a otra.

class CuentaBancaria:
    def __init__(self, saldo):
        self.__saldo = saldo

    def obtener_saldo(self):
        return self.__saldo

    def depositar(self, cantidad):
        self.__saldo += cantidad

    def retirar(self, cantidad):
        self.__saldo -= cantidad

    def transferir(self, cuenta_destino, cantidad):
        self.retirar(cantidad)
        cuenta_destino.depositar(cantidad)

cuenta1 = CuentaBancaria(1000)
cuenta2 = CuentaBancaria(500)
print(cuenta1.obtener_saldo())
print(cuenta2.obtener_saldo())
cuenta1.transferir(cuenta2, 200)
print(cuenta1.obtener_saldo())
print(cuenta2.obtener_saldo())
```
